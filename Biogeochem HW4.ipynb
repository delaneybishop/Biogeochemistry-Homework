```python
#Part 1 Figure Wheat Inorganic P

#Define method characteristics 
method = ['Resin','NaHCO3', 'CHCl-released NaHCO3', 'NaOH', 
        'Sonicate-NaOH', 'HCl'] 

#Define the data
data = [Wheat_Resin_inorg_percent,Wheat_NaHCO3_inorg_percent,Wheat_CHCl_NaHCO3_inorg_percent,
       Wheat_NaOH_inorg_percent, Wheat_Sonicate_NaOH_inorg_percent, Wheat_HCl_inorg_percent]


#Create the color parameters for each pool
color = ("orange", "cyan", "red", "pink", "gold", "greenyellow")

#Define each pool wedge property
wp = { 'linewidth' : 1, 'edgecolor' : "black", "antialiased" : True }
explode = (0, 0.1, 0.2, 0.1, 0.1, 0.1)
#Create autocpt arguments
#Function used to label the wedges with the numeric value from the defined data
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n".format(pct, absolute)

#Create pie chart
fig, ax = plt.subplots(figsize =(10, 6))
wedges, texts, autotexts = ax.pie(data, 
                                  autopct = lambda pct: func(pct, data),
                                  colors = color,
                                  explode=explode,
                                  startangle = 90, #Rotate by 90 degrees counter clockwise 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="black")) #Define the text within the wedges to be black

#Create the legend
ax.legend(wedges, method, title ="Inorganic Phosphorus Fractions", loc ="upper left", bbox_to_anchor =(1, 0, 0.5, 1))

#Determine size and define bold text within the pie chart
plt.setp(autotexts, size = 10, weight ="bold")
plt.savefig('Wheat_Inorg_Phosphorus_Fractions.png')
plt.show()
```

```python
#Part 2 calculation

#Total P in kg for the pasture grazing management and the wheat harvesting management

#Additional information
#P wheat = μg P/g soil 
#P pasture = μg P /g soil 
#pasture bulk density = 1.4 g/cm3
#wheat bulk density = 1.6 g/cm3
#1 μg = 10**-9 kg
#Area_acre = 2 acres or 8093.72 m2 
#Area_m3 = 8093.72 m2*1 m = 8093.72 m3
#m3 = 1000000 cm3
#Area_cm3 = 8093.72 m3*1000000 = 8093720000 cm3

#Compute total P for wheat and pasture by adding the values for each pool and method
P_wheat = 23.8+15.4+2.8+20.3+7.8+117+12.3+2.2+73+15+261
P_pasture = 41.2+29.7+3.8+23+12.6+133+19.4+4.1+107+22+376

#Convert μg P/g soil to kg P/g soil 
P_wheat_kgP_gsoil = P_wheat*(10**-9)
P_pasture_kgP_gsoil = P_pasture*(10**-9)

#Multiply P kg wheat and P kg pasture by bulk density
P_wheat_kg_bd = P_wheat_kgP_gsoil*1.6
P_pasture_kg_bd = P_pasture_kgP_gsoil*1.4

#Multiply by 8093720000 cm3 to conver to kg P
P_wheat_kg = P_wheat_kg_bd*8093720000
P_pasture_kg = P_pasture_kg_bd*8093720000

#display total P in kg for each management type
display('Total wheat harvesting method Phosphorus in kg', P_wheat_kg)
display('Total pasture grazing method Phosphorus in kg', P_pasture_kg)
```
